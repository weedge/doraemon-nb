# 总结
- 中文支持不好，不过可以用中文音频数据进行微调。
- 流式处理默认使用28 tokens to gen waveform , no overlap
- 提供的orpheus_tts是结合vllm进行流式输出， https://github.com/canopyai/Orpheus-Speech-PyPi/blob/main/orpheus_tts/decoder.py#L105C1-L110C1

首次28 tokens 后续每7个token + overlap 21 tokens 流式生成waveform
